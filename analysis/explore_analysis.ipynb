```python


def plot_model_accuracy_scatter(df: pd.DataFrame, model_name: str):
    """
    Single scatter plot for one model.

    x-axis: dataset/task
    y-axis: accuracy
    color: strategy
    """

    model_df = df[df["model"] == model_name].copy()

    tasks = ["arc", "gsm8k", "humaneval", "truthfulqa"]
    strategies = ["single_pass", "best_of_n", "self_refine", "oracle"]

    colors = {
        "single_pass": "#d62728",
        "best_of_n": "#1f77b4",
        "self_refine": "#2ca02c",
        "oracle": "#9467bd"
    }

    fig = go.Figure()

    for strategy in strategies:

        strat_df = model_df[model_df["strategy"] == strategy]

        fig.add_trace(
            go.Scatter(
                x=strat_df["task"],
                y=strat_df["accuracy_mean"],
                mode="markers",
                name=strategy,
                opacity = 1.0 if strategy == "single_pass" else 0.5,
                marker=dict(
                    size = 14 if strategy == "single_pass" else 10,
                    color=colors[strategy]
                ),

            )
        )

    fig.update_layout(
        title=f"Accuracy Across Datasets — {model_name}",
        xaxis_title="Dataset",
        yaxis_title="Accuracy",
        yaxis=dict(range=[0.0, 1.05]),
        width=900,
        height=500,
        legend_title="Strategy",
        title_x=0.5,
        font=dict(
            family="Aptos",
            size=16,
            color="black"
        )
    )

    fig.update_yaxes(
        showgrid=True,
        gridwidth=1,
        gridcolor="rgba(0,0,0,0.2)",
        zeroline=False
    )

    return fig


fig = plot_model_accuracy_scatter(df_master, model_name="claude")
fig.show()
```
